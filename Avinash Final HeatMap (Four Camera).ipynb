```python
import numpy as np
import cv2
import copy
```

```python
cap = cv2.VideoCapture('input_part1.mp4')
cap2 =cv2.VideoCapture('input_part2.mp4')
cap3 =cv2.VideoCapture('input_part3.mp4')
cap4 =cv2.VideoCapture('input_part4.mp4')
```

```python
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
if (cap2.isOpened()== False): 
  print("Error opening video stream or file")
if (cap3.isOpened()== False): 
  print("Error opening video stream or file")
if (cap4.isOpened()== False): 
  print("Error opening video stream or file")   
```

```python
first_iteration_indicator = 1
background_subtractor = cv2.bgsegm.createBackgroundSubtractorMOG()
i = 0
images1 = []
images2 = []
final_stitched = []
```

```python
while(cap.isOpened() & cap2.isOpened() ):
    # Capture frame-by-frame
    ret, frame = cap.read()
    ret, frame1 = cap2.read()
    ret, frame2 = cap3.read()
    ret, frame3 = cap4.read()
  
    images1.append(frame)
    images1.append(frame1)
    images2.append(frame2)
    images2.append(frame3)
  
    if ret == True:
                      
        #stitcher = cv2.Stitcher.create()
        stitcher = cv2.Stitcher_create()
        (status1, stitched1) = stitcher.stitch(images1)
        
           
        if(status1 == cv2.STITCHER_OK):
            print("Stitcher1 OK")
        else:
            print("Stitcher1 Not OK")
            
        (status2, stitched2) = stitcher.stitch(images2)    
          
            
        if(status2 == cv2.STITCHER_OK):
            print("Stitcher2 OK")
        else:
            print("Stitcher2 Not OK")    
            
        final_stitched.append(stitched1)    
        final_stitched.append(stitched2) 
        
        (final_status, stitched) = stitcher.stitch(final_stitched)    
         
        if(final_status == cv2.STITCHER_OK):
            print("Final Stitcher OK")
        else:
            print("Final Stitcher Not OK")      
            
            
        frame_main = stitched.copy()
        images1 = []
        images2 = []
        final_stitched = []
        
                
        # If first frame
        if first_iteration_indicator == 1:
            first_frame = copy.deepcopy(frame_main)
            #cv2.imwrite("First_frame_four_camera.jpg",first_frame)
            height, width = frame_main.shape[:2]
            accum_image = np.zeros((height, width), np.uint8)
            first_iteration_indicator = 0
        else:
            
            frame_main = cv2.resize(frame_main, (accum_image.shape[1],accum_image.shape[0]),cv2.INTER_LINEAR)
            
            filter = background_subtractor.apply(frame_main)  # remove the background
                            
            #cv2.imwrite('filter_copy_four_camera.jpg', filter)

            threshold = 2
            maxValue = 2
            ret, th1 = cv2.threshold(filter, threshold, maxValue, cv2.THRESH_BINARY)
           
                                 
            # add to the accumulated image
            accum_image = cv2.add(accum_image, th1)
           
            #cv2.imwrite('mask_copy_four_camera.jpg', accum_image)
         
            # color map applied to mask
           
            color_image_video = cv2.applyColorMap(accum_image, cv2.COLORMAP_HOT)
            
            video_frame = cv2.addWeighted(frame_main, 0.7, color_image_video, 0.7, 0)
            
            i+=1
            
            print("i:",i)  
            
            #cv2.imwrite("Final_frame_four_camera.jpg", video_frame)
             
            cv2.namedWindow("Heatmap video",cv2.WINDOW_NORMAL)  
            cv2.imshow("Heatmap video",video_frame)   
            
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break



cap.release()
cap2.release()
cap3.release()
cap4.release()

cv2.destroyAllWindows()     
```
